```python
# Importing libraries

from bs4 import BeautifulSoup
import requests

```

```python
# Adding URL

link = "https://www.pcworld.com/article/2105289/this-rock-solid-brother-wi-fi-laser-printer-is-just-70.html"
data = requests.get(link)

print(data)
```

```python
# Creating Soup object

soup = BeautifulSoup(data.text, "html.parser")

#print (soup)
#or
print(soup.prettify())

```

```python
# Retrieve the Data


price = soup.find_all(class_="entry-title")[0]

price_text= price.find("h1")



print(price_text)



for i in price:
     print(i.text)
```

```python
#  Retrieve the Data ( image )

img_tag = soup.find_all("img")

print(img_tag)
```
